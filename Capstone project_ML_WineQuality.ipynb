## Predict the quality of the wine

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
df_wine_1
```

```python
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']
x = df_wine_1[features]
y = df_wine_1['quality']
```

```python
#plotting features vs quality
sns.pairplot(df_wine_1,x_vars=features,y_vars='quality',kind='reg',size=5,aspect=0.5)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=99)
```

```python
x_train
```

```python
x_test
```

```python
lr = LinearRegression(fit_intercept=False)
lr.fit(x_train,y_train)
```

```python
accuracy = lr.score(x_test, y_test)
"Accuracy: {}%".format(int(round(accuracy * 100)))
```

## Cluster wine sample by its quality score

```python
from sklearn.cluster import DBSCAN,KMeans
```

```python
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']
x_4 = df_wine_1[features]
y_4 = df_wine_1['quality']
```

```python
x_4=np.array(x_4)
x_4
```

```python
sklearn_kmeans = KMeans(n_clusters=6, random_state=399)
sklearn_kmeans
```

```python
sklearn_kmeans.fit(x_4)
```

```python
sklearn_kmeans.labels_
```

```python
def visualize_clusters(X, cluster_preds):
    colors = ['r', 'g', 'b', 'k', 'c', 'y','w','purple','pink','orange','yellow','brown']
    for x, cluster in zip(X, cluster_preds):
        plt.plot(x[0], x[1], f'{colors[cluster]}o')
    plt.show()
```

```python
visualize_clusters(x_4, sklearn_kmeans.labels_)
```

```python
sklearn_dbscan = DBSCAN(
    eps=5.55,
    min_samples=55
)
```

```python
sklearn_dbscan.fit(x_4)
```

```python
sklearn_dbscan.labels_
```

```python
def visualize_clusters_1(X, cluster_preds):
    colors = ['r', 'g', 'b', 'k', 'c', 'y']
    for x, cluster in zip(X, cluster_preds):
        plt.plot(x[0], x[1], f'{colors[cluster]}o')
    plt.show()
```

```python
visualize_clusters_1(x_4, sklearn_dbscan.labels_)
```
